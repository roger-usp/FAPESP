```python
import pandas as pd
import sqlite3
import numpy as np
```

```python
def get_cod(mun_name, uf_name):
    con = sqlite3.connect("../FAPESP.db")
    my_cursor = con.cursor()
    result = my_cursor.execute(
        """
        SELECT municipios.id
        FROM municipios
        INNER JOIN imediatas ON municipios.imed_id = imediatas.id
        INNER JOIN intermediarias ON imediatas.inter_id = intermediarias.id
        INNER JOIN uf ON intermediarias.uf_id = uf.id
        WHERE municipios.name = ? AND uf.name = ?
        """,
        (mun_name, uf_name)
    )
    return list(result)[0][0]
```

```python
def add_cod(row):
    try:
        row["mun_id"] = get_cod(row["Local"], row["UF"])
    except:
        print(row)
    return row

def format_table(df):
    df = df.copy()
    df = df.loc[df["Tipo Região"] == "Município"]
    df = df[["Local", "UF", "Quantidade produzida (Toneladas)"]]
    df = df.apply(add_cod, axis=1)
    df["Quantidade produzida (Toneladas)"] = df["Quantidade produzida (Toneladas)"].apply(lambda x: pd.to_numeric(x, errors="coerce"))
    df = df.dropna(subset=["Quantidade produzida (Toneladas)"])
    return df
```

```python
cafe = pd.read_excel("PAM/2021 Café (em grão) Total.xlsx")
laranja = pd.read_excel("PAM/2021 Laranja.xlsx")
soja = pd.read_excel("PAM/2021 Soja (em grão).xlsx")
```

```python
cafe = format_table(cafe)
cafe
```

```python
laranja = format_table(laranja)
laranja
```

```python
soja = format_table(soja)
soja
```

```python
soja.to_excel("soja.xlsx", index=False)
laranja.to_excel("laranja.xlsx", index=False)
cafe.to_excel("cafe.xlsx", index=False)
```
